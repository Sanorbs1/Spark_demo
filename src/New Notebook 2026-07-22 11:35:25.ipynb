```python

df = spark.table("worry_the_killer.killer.sales")
df.show()
df.printSchema()
```

```python
from pyspark.sql.functions import col, sum, avg, when, count

df_clean = df.na.fill({"amount": 0})

high_sales = df_clean.filter(col("amount") > 100)
high_sales.show()

summary_df = df_clean.groupBy("region").agg(
    sum("amount").alias("total_amount"),
    avg("amount").alias("avg_amount")
)
summary_df.show()
```

```python
summary_df.write.mode("overwrite").saveAsTable("worry_the_killer.killer.sales_summary")
```

```python
df = spark.table("worry_the_killer.killer.sales")

from pyspark.sql.functions import col, countDistinct, spark_partition_id

df2 = df.withColumn("partition_id", spark_partition_id())
partition_count = df2.select(countDistinct("partition_id").alias("num_partitions"))
partition_count.show()
```

```python
from pyspark.sql.functions import col, sum

df_clean = df.na.fill({"amount": 0})
result = df_clean.groupBy("region").agg(sum("amount").alias("total_amount"))
result.show()
```

```python
spark.sql("
USE CATALOG worry_the_killer;"
)

spark.sql("""
SELECT region,
       SUM(amount) AS total_amount,
       AVG(amount) AS avg_amount
FROM sales
GROUP BY region
""").show()
```
