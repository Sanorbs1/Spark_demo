```sql
USE CATALOG worry_the_killer;
USE SCHEMA killer;

CREATE TABLE IF NOT EXISTS killer.sales (
  id INT,
  region STRING,
  amount DOUBLE
);

INSERT INTO killer.sales VALUES
(1, 'APAC', 120.0),
(2, 'EMEA', 80.0),
(3, 'APAC', 300.0),
(4, 'AMER', 150.0),
(5, 'EMEA', NULL);
SELECT region,
       SUM(amount) AS total_amount,
       AVG(amount) AS avg_amount
FROM sales
GROUP BY region
```
